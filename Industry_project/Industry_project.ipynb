# Проект: промышленность

*Чтобы оптимизировать производственные расходы, металлургический комбинат «Стальная птица» решил уменьшить потребление электроэнергии на этапе обработки стали. Для этого комбинату нужно контролировать температуру сплава. **Ваша задача — построить модель, которая будет её предсказывать**.*

# Описание этапа обработки

*Сталь обрабатывают в металлическом ковше вместимостью около 100 тонн. Чтобы ковш выдерживал высокие температуры, изнутри его облицовывают огнеупорным кирпичом. Расплавленную сталь заливают в ковш и подогревают до нужной температуры графитовыми электродами. Они установлены в крышке ковша.* 

*Из сплава выводится сера (этот процесс — десульфурация), добавлением примесей корректируется химический состав и отбираются пробы. Сталь легируют — изменяют её состав — подавая куски сплава из бункера для сыпучих материалов или проволоку через специальный трайб-аппарат (от англ. tribe — «масса»).*

*Перед тем как первый раз ввести легирующие добавки, измеряют температуру стали и производят её химический анализ. Потом температуру на несколько минут повышают, добавляют легирующие материалы и продувают сплав инертным газом. Затем его перемешивают и снова проводят измерения. Такой цикл повторяется до достижения целевого химического состава и оптимальной температуры плавки.*

*Тогда расплавленная сталь отправляется на доводку металла или поступает в машину непрерывной разливки. Оттуда готовый продукт выходит в виде заготовок-слябов (от англ. slab — «плита»).*

```python
# функция для анализа по гистограммам и ящикам с усами
float_type = data_arc['active_power'].dtype

def histogram(data):
    display(data.drop(['key'], axis = 1).describe().T)
    for i in data.columns:
        if data[i].dtype == float_type:
            
            # гистограмма
            fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
            ax2.hist(data[i], bins=15, color='orange', edgecolor='black')
            ax2.set_xlim(data[i].min(), data[i].max() + 0.5)
            ax2.set_title(f'График распределения {i}')
            ax2.set_xlabel(i)
            ax2.set_ylabel('Частота')
            
            # Ящик с усами
            ax1.boxplot(data[i], vert=False)
            ax1.set_title(f'Ящик с усами для {i}')
            ax1.set_xlabel(i)

            plt.tight_layout() 
            plt.show()
```

*Так как наша задача - оптимизировать количество электроэнергии, то нужно создать такой столбец. А чтобы его создать, нужно провести несколько этапов:*
- 1) посчитаем полную мощность (корень из сумм квадратов активной и реактивной мощностей)
- 2) посчитаем длительность нагрева дугой (разность между концом и началом нагрева)
- 3) посчитаем затраты электроэнергии (перемножим длительность на полную мощность)

```python
# считаем полную мощность 
data_arc['full_power'] = (data_arc['active_power'] ** 2 + data_arc['reactive_power'] ** 2) ** 0.5

# посчитаем длительность нагрева дугой
data_arc['arc_heating_duration'] = (data_arc['end_arc_heating'] - data_arc['start_arc_heating']).astype('timedelta64[s]')
data_arc['arc_heating_duration'] = data_arc['arc_heating_duration'].astype('int')

# посчитаем затраты электроэнергии 
data_arc['electricity_costs'] = data_arc['full_power'] * (data_arc['arc_heating_duration'])

display(data_arc.head(10))
```

*теперь нам не нужны столбцы начала нагрева, конец, активная и реактивная мощности, удалим их*

```python
data_arc = data_arc.drop(['start_arc_heating', 'end_arc_heating', 'active_power', 'reactive_power'], axis = 1)
display(data_arc.head())
```

*Ранее заметил, что в каких то датафреймах даются толькло одни данные для одной партии, а в каких то нет, поэтому там, где нет - сложим энергозатраты, мощность и время, чтобы были только одни данные для одной партии*

```python
arc_func = {'key' : 'count',
           'full_power' : 'sum',
           'arc_heating_duration' : 'sum',
           'electricity_costs' : 'sum'}
data_arc = data_arc.groupby('key').agg(arc_func)
data_arc = data_arc.rename(columns = {'key' : 'number_iterations'})
display(data_arc.head(10))
```

## Подготовка данных

```python
RANDOM_STATE = 241224
TEST_SIZE = 0.25
```

```python
X = general_data.drop('temp', axis = 1)
y = general_data['temp']

X_train, X_test, y_train, y_test = train_test_split(X,
                                                   y,
                                                   test_size = TEST_SIZE,
                                                   random_state = RANDOM_STATE)
```

```python
scaler = StandardScaler()

X_train_scaler = scaler.fit_transform(X_train)
X_test_scaler = scaler.transform(X_test)

X_train_scaler = pd.DataFrame(X_train_scaler, columns = X_train.columns)
X_test_scaler = pd.DataFrame(X_test_scaler, columns = X_test.columns)
```

*Данные подготовили, разбили на 3 выборки, чтобы проверить mae на валидационной выборке. Теперь начинаем обучение*

## Обучение моделей машинного обучения

*Обучим 3 модели:*
- LinearRegression
- CatBoost
- DecisionTreeClassifier

*Сравним их по метрике MAE*

###  LinearRegression

```python
model_lr = LinearRegression()
mae_lr = abs(cross_val_score(model_lr, X = X_train_scaler, y = y_train, cv = 3, 
                        scoring='neg_mean_absolute_error').mean())
print(mae_lr)
```

### CatBoost

```python
model_cbr = CatBoostRegressor(random_state=RANDOM_STATE)
parameters = [{'learning_rate':[0.05, 0.06, 0.07, 0.08, 0.09], 'verbose':[False]}]
model_cbr = GridSearchCV(model_cbr, parameters, scoring='neg_mean_absolute_error', cv = 3)
model_cbr.fit(X_train, y_train)
print(model_cbr.best_params_)
```

```python
mae_cbr = abs(model_cbr.best_score_)
print(mae_cbr)
```

### DecisionTreeClassifier

```python
model_dtc = DecisionTreeClassifier(random_state=RANDOM_STATE)

param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [None] + list(range(1, 11)),
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4, 5],
}

grid_search = GridSearchCV(estimator=model_dtc,
                           param_grid=param_grid,
                           scoring='neg_mean_absolute_error',  
                           cv=3,  
                           verbose=1,
                           n_jobs=-1)  

grid_search.fit(X_train, y_train)

best_params = grid_search.best_params_


print("Лучшие гиперпараметры:", best_params)
```

```python
mae_dtc = abs(grid_search.best_score_)

print(mae_dtc)
```

```python
model_metrics = pd.DataFrame(data = {'mae' : [mae_lr, mae_cbr, mae_dtc]}, 
                            index = ['LinearRegression', 'CatBoost', 'DecisionTreeClassifier'])
display(model_metrics)
```

**MAE у LinearRegression самое наименьшее - 6.23, она и переходит к проверке на тестовой выборке!**
